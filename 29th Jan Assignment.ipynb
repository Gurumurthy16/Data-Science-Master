Q1: Python Programming Language was developed by **Guido van Rossum**.

Q2: Python supports **multi-paradigm programming**. It primarily supports **procedural**, **object-oriented**, and **functional** programming. It is known for its simplicity and readability.

Q3: Yes, Python is **case-sensitive** when dealing with identifiers. This means that variables or function names with different cases (e.g., `myVariable` and `myvariable`) are treated as distinct entities.

Q4: The correct extension for a Python file is **".py"**. For example, "my_script.py" is a valid Python file name.

Q5: Python is an **interpreted** language. Python code is executed line by line by the Python interpreter without the need for a separate compilation step.

Q6: In Python, you can define code blocks using the following:
   - **Functions**: Defined using the `def` keyword.
   - **Conditional Statements**: Such as `if`, `elif`, and `else` for branching logic.
   - **Loops**: Including `for` and `while` loops for iteration.
   - **Classes**: Defined using the `class` keyword to create objects and encapsulate behavior.

Q7: The character **"#"** is used to give **single-line comments** in Python. Anything following a "#" on a line is treated as a comment and is not executed.

Q8: To find the version of Python you are currently working on, you can use the `sys` module's `sys.version` attribute or the `platform` module's `platform.python_version()` function.

```python
import sys
import platform

print("Python version (sys):", sys.version)
print("Python version (platform):", platform.python_version())
```

Q9: Python supports the creation of **anonymous functions** at runtime using a construct called **"lambda"**. Lambda functions are also known as **anonymous or inline functions**.

Q10: **PIP** stands for **Python Package Index**. It is a package manager for Python that allows you to install, upgrade, and manage Python packages and libraries.

Q11: Some built-in functions in Python include `print()`, `len()`, `range()`, `input()`, `type()`, `int()`, `str()`, `float()`, and many more.

Q12: The maximum possible length of an identifier (variable or function name) in Python is **unlimited**. However, it is recommended to keep identifiers reasonably short and meaningful for code readability.

Q13: Benefits of using Python include:
   - **Readability**: Python's clean and easy-to-read syntax.
   - **Versatility**: Supports multiple programming paradigms.
   - **Large Standard Library**: A wide range of built-in modules and libraries.
   - **Community and Support**: Active and large user community.
   - **Cross-Platform**: Runs on various platforms.
   - **Open Source**: Python is freely available and can be customized.
   - **Diverse Applications**: Used in web development, data analysis, machine learning, scripting, and more.

Q14: Memory in Python is managed using a combination of **automatic memory management** (via reference counting) and a **garbage collector** that reclaims memory that is no longer in use.

Q15: To install Python on Windows and set path variables, follow these steps:
   1. Download the Python installer from the official Python website.
   2. Run the installer and select the option to add Python to the system PATH during installation.
   3. Complete the installation process.
   4. Open a command prompt and type `python` to verify the installation.

Q16: Yes, **indentation is required** in Python. Indentation is used to define code blocks and is a fundamental part of Python's syntax. It is used to group statements within functions, loops, and conditional blocks.
